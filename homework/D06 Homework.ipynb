2. 請對一個 5x5 的隨機矩陣作正規化的操作。


````python
A = np.random.random((5, 5))
print(A)
A_min, A_max = A.min(),A.max()
A = (A - A_min) / (A_max - A_min)

print(A)
````

3. 請建立一個長度等於 10 的正整數向量，並且將其中的最大值改成 -1。


````python
B = np.random.randint(1,10,size=10)
print(B)
print(B.max())

B[B==(B.max())] = -1
````
